# Extraction

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
scoreArray = []
numberArray = []

def read_all_files(line_number):
    folder_path = "Sequence//"


    
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
    
        if os.path.isfile(file_path):
            try:
                with open(file_path, 'r') as file:
                    for i, line in enumerate(file):
                        if i + 1 == line_number:
                            number = int("".join(filter(str.isdigit, filename)))
                            numberArray.append(number)
                            score = float(line.split()[3])
                            scoreArray.append(score)
                            break
            except FileNotFoundError:
                return None
```

# Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ecnodes the letters going from 1 to 20
amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
aa_to_idx = {aa: i + 1 for i, aa in enumerate(amino_acids)}  # Start at 1 to reserve 0 for padding

def encode_sequence(seq):
    return [aa_to_idx[aa] for aa in seq if aa in aa_to_idx]
```

```python
# Encoding and padding
max_len = max(data["Sequences"].apply(len))
encoded_seqs = [encode_sequence(seq) for seq in data["Sequences"]]
padded_seqs = [seq + [0] * (max_len - len(seq)) for seq in encoded_seqs]

X = torch.tensor(padded_seqs, dtype=torch.long)
y = torch.tensor(data["Scores"].values, dtype=torch.float32)
```

```python
# Data laoder
class SequenceDataset(Dataset):
    def __init__(self, X, y):
        self.X = X # Encoded Sequences
        self.y = y # Target Scores

    def __len__(self):
        return len(self.X) # How many samples

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx] # Grab single sequence

dataset = SequenceDataset(X, y) # Creates data set with features x and y
dataloader = DataLoader(dataset, batch_size=2, shuffle=True) # 2 samples at a time
```

```python
# Model
class LSTMRegressor(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0) # Creates embedding layer of vectors of size embed_dim
        self.lstm = nn.LSTM(embed_dim, hidden_dim, num_layers=2, batch_first=True, dropout=0.3)
        #self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        x = self.embedding(x)
        _, (h_n, _) = self.lstm(x)
        out = self.fc(h_n[-1])
        return out.squeeze(1)

vocab_size = len(amino_acids) + 1  # +1 for padding
model = LSTMRegressor(vocab_size=vocab_size, embed_dim=32, hidden_dim=128) # Creates the model with embed size of 32 and hidden size 128
#model = LSTMRegressor(vocab_size=vocab_size, embed_dim=10, hidden_dim=64)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Loop to train
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch_X, batch_y in dataloader:
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {total_loss/len(dataloader):.4f}")
```

```python
# Cell 8: Predictions and visualization
model.eval()
with torch.no_grad():
    predictions = model(X).numpy()

# Scatter plot: Predicted vs Actual
plt.figure(figsize=(6, 6))
plt.scatter(y.numpy(), predictions, alpha=0.5)
plt.plot([-13, -9], [-13, -9], '--r')
plt.xlabel("Actual Score")
plt.ylabel("Predicted Score")
plt.title("Predicted vs Actual")
plt.grid(True)
plt.show()
```
