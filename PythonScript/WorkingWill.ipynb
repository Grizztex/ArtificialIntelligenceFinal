##Analyzing data from pdbqt files to extract score:
1. read file
2. file if match:
"REMARK VINA RESULT:   -11.417      5.250     23.376"
3. Extract first number
4. store in a list for each sequences (9 models in each sequence) 
5. Calculate average
6. TAble for each sequences: sequence ID, model, result

```python
in_dir = r"C:\Users\daott\Downloads\Insulin Aggregation\Docking Vina\TestScriptScore"
files = os.listdir(in_dir)
pattern = r"REMARK VINA RESULT:[ ]+(-\d+.\d+).*"
table = pd.DataFrame(columns = ["Model 1","Model 2","Model 3","Model 4","Model 5","Model 6","Model 7","Model 8","Model 9"],dtype = float)
for file in files:
    ID =  int(file.split('_')[0][8:])##Extract number out of sequence
    print(ID)
    path = in_dir+"\\" + file
    resultList = []
    with open(path, "r") as f:
        for line in f:
            if re.match(pattern,line):#If the line containing result
                result = float(re.search(pattern,line).group(1))
                resultList.append(result)
    ##If less than 9 poses were found, append the missing value with 0
    for x in np.arange(9-len(resultList)):
        resultList.append(0)
    table.loc[ID,:] = resultList
table 
```

```python
##replace 0 with NaN
table2 = table[table<0]
table2["Average"] = table2.mean(axis=1)
table2
```
