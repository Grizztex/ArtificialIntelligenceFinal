```python
from pymol import cmd
import pandas as pd
import numpy as np
import os
```

```python
# Define the directory where the files are located
input_dir = r"C:\Users\tiend\Downloads\Insulin Aggregation\Alphafold Result\result(1-267)_1st_rank"  # Replace with the directory containing the input files
output_dir = r"C:\Users\tiend\Downloads\Insulin Aggregation\Alphafold Result\toPdb"  # Replace with the directory for saving output PDB files


# Range of sequences (e.g., sequence1 to sequence1000)
sequence_range = range(1, 267)  # Adjust this range based on how many sequences you have

# Loop through the sequence numbers
for fileName in os.listdir(r"C:\Users\tiend\Downloads\Insulin Aggregation\Alphafold Result\result(1-267)_1st_rank"):
    input_filepath = os.path.join(input_dir, fileName)  # Assuming the files are PDB format
    # Load the structure file
    cmd.load(input_filepath)
    saveFileName = fileName.split("-")[0]
    # Save the structure with the same filename in the output directory
    output_filepath = os.path.join(output_dir, saveFileName)
    cmd.save(output_filepath)
    
    # Remove the loaded structure from PyMOL after saving
    cmd.delete("all")
```
